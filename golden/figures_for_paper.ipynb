```python
spec = pickle.load(open(base + '/extracted/specs/mel/CR05EF-13527_20130916_0823.pkl'))
spec = np.log(spec * 10 + 0.001)
plt.figure(figsize=(15, 7))
plt.imshow(spec[::-1, 280:380], cmap=plt.get_cmap('inferno'))
plt.axis('off')
plt.savefig('../paper/figs/spec.png')
```

```python
S = spec.copy()#[::-1, 280:380] 
S -= np.median(S, axis=1, keepdims=True)
S -= S.mean()
S /= S.var()
plt.figure(figsize=(15, 7))
plt.imshow(S[::-1, 280:380], cmap=plt.get_cmap('inferno'))
plt.axis('off')
plt.savefig('../paper/figs/spec_normed.png')
```
