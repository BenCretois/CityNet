```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import yaml
```

```python
strategies = ['None', 'stowell_half', 'stowell_half_rescale', 'stowell_full',
              'overall_median', 'overall_median_rescale', 'sum_to_one',
              'equal_power', 'local_normalisation', 'simple_whiten']
```

```python
base = '../results/normalisation_run_0000/'
res = {}
for strat in strategies:
    path = base + 'split_2_' + strat + '/' + 'best_and_final.yaml'
    res[strat] = yaml.load(open(path))
    
```

```python
final_accs = np.array([res[strat]['final_val_acc'] for strat in strategies])
idxs = np.argsort(final_accs)
plt.barh(np.arange(len(final_accs)), final_accs[idxs])
plt.gca().set_yticks(np.arange(len(strategies))+0.5)
plt.gca().set_yticklabels([strategies[idx] for idx in idxs]);
```
